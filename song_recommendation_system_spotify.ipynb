```python
import pandas as pd
import numpy as np
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
songs = pd.read_csv('spotify_millsongdata.csv')
```

```python
songs
```

```python
df = songs.drop('link', axis=1).sample(10000, random_state=101)
```

```python
df.shape
```

```python
df = df.reset_index().drop('index', axis=1)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.text[0]
```

```python
df.text[0].replace('\r\n', '').replace("\'", '')
```

```python
 df.text =  df.text.str.replace('\r\n', ' ').str.replace(',', ' ').str.replace('?', '').str.replace("\'", ' ').str.replace('"', '')
```

```python
df.text[0]
```

```python
df.head()
```

```python
df.artist.value_counts()
```

```python
df.text = df.text.str.split(' ')
```

```python
df.head()
```

```python
ps = PorterStemmer()
```

```python
def clean_text(text_list):
    new_list = []
    for word in text_list:
        if word != '':
            new_list.append(ps.stem(word))
    return new_list
```

```python
df.text = df.text.apply(clean_text)
```

```python
df.text[0]
```

```python
df.head()
```

```python
df.text = df.text.apply(lambda x: ' '.join(x))
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english', analyzer='word' , max_features=5000)
```

```python
metrix = tfidf.fit_transform(df.text)
```

```python
metrix.toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
song_simi = cosine_similarity(metrix)
```

```python
song_simi
```

```python
def spotify_song_recommender(song_name):
    idx = df[df['song']==song_name].index[0]
    songs_list = sorted(enumerate(song_simi[idx]), reverse=True, key = lambda x:x[1])[1:6]
    song_idx_list =[]
    for i, j in songs_list:
        song_idx_list.append(i)
    return song_idx_list
```

```python
spotify_song_recommender('Alabama Sky')
```
